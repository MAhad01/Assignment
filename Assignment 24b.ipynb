```python
# Ans 1

# Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. 
# It is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack. 
# Matplotlib makes easy things easy and hard things possible1. It produces publication-quality figures in a variety of hardcopy formats and interactive environments across platforms.

# Matplotlib is used for creating various types of plots such as:

# Line plot
# Scatter plot
# Bar plot
# Histogram
# Pie chart
```

```python
# Ans 2

# A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. 
# It is used to observe and visually display the relationship between variables. The values of the variables are represented by dots. 
# Scatter plots are often used when you want to assess the relationship between two variables. 
# For example, you might use a scatter plot to determine whether there is a relationship between a person’s height and weight, or between the age of a car and its selling price. 
# Scatter plots can also be used to identify outliers in data sets.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# Ans 3

# The subplot() function is used to create multiple plots in a single figure. 
# It is a convenient way to create subplots of different sizes and arrange them on a single grid. 
# The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the subplot.

# Here are four line plots using the subplot() function with the data you provided:

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('Line Plot - y1 vs x1')

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('Line Plot - y2 vs x2')

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('Line Plot - y3 vs x3')

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('Line Plot - y4 vs x4')

plt.suptitle('Four Line Plots using Subplot() Function')
plt.show()
```

```python
# Ans 4

# A bar plot is a type of plot or mathematical diagram that represents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. 
# It is used to compare different categories of data. The bars can be plotted vertically or horizontally. 
# Bar plots are often used when you want to compare the values of different categories of data. 
# For example, you might use a bar plot to compare the sales of different products in a store, or the number of students in different classes in a school.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
```

```python
# Ans 5

# A box plot is a type of plot or mathematical diagram that displays the distribution of a dataset based on five summary statistics: minimum, 
# first quartile, median, third quartile, and maximum. It is used to identify the range and distribution of a dataset, as well as to identify any outliers. 
# Box plots are often used when you want to compare the distribution of different datasets.

import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig = plt.figure(figsize =(10, 7))

# Creating axes instance
ax = fig.add_subplot(111)

# Creating axes instance
bp = ax.boxplot(data)

# Changing color and linewidth of whiskers
for whisker in bp['whiskers']:
    whisker.set(color ='#8B008B', linewidth = 1.5,
                linestyle =":")
 
# Changing color and linewidth of caps
for cap in bp['caps']:
    cap.set(color ='#8B008B', linewidth = 2)
 
# Changing color and linewidth of medians
for median in bp['medians']:
    median.set(color ='red', linewidth = 3)
 
# Changing style of fliers
for flier in bp['fliers']:
    flier.set(marker ='D', color ='#e7298a',
              alpha = 0.5)
     
# X-Axis labels
ax.set_xticklabels(['Box1', 'Box2'])
 
# Adding title
plt.title("Boxplot using Matplotlib")
 
# Removing top axes and right axes ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()
     
# Show plot
plt.show()
```
