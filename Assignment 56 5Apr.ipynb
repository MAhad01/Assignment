```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Ans 1
df=pd.read_csv('diabetes.csv')
```

```python
df.isnull().sum()
# No value is missing
```

```python
plt.boxplot(df)
```

```python
# Ans 2
df['Insulin']= np.where(df["Insulin"] >250, 115,df['Insulin'])
# since insuline has max outliers, we will replace all values larger than 250 by mean
```

```python
# repla cing null vaues in Blood Pressure, Insuline, Skin Thickness and BMI with their mean
df['Insulin']= np.where(df["Insulin"] ==0, 115,df['Insulin'])
df['BloodPressure']= np.where(df["BloodPressure"] == 0, 80,df['BloodPressure'])
df['BMI']= np.where(df["BMI"] == 0, 21,df['BMI'])
```

```python
df['SkinThickness']= np.where(df["SkinThickness"] == 0, 32,df['SkinThickness'])
df['Glucose']= np.where(df["Glucose"] == 0, 120,df['Glucose'])
```

```python
df.describe()
```

```python
df.info()
```

```python
# Ans 3

x=df.iloc[:, :-1]
y=df.iloc[:, -1]
```

```python
x
```

```python
y
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y, test_size=20, random_state=42)
```

```python
X_train, X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
```

```python
classifier.fit(X_train, y_train)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# graphical representation of decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier, filled=True)
```

```python
# Ans 4
# Create a decision tree classifier with the ID3 algorithm
clf = DecisionTreeClassifier(criterion='entropy')
```

```python
# Define the hyperparameters to search over
param_grid = {
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
# Use cross-validation to optimize the hyperparameters
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(clf, param_grid, cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
# get best hyperparameters
best_param=grid.best_params_
```

```python
best_param
```

```python
# Train the decision tree model using the best hyperparameters

clf2=DecisionTreeClassifier(criterion= 'gini', max_depth= 5, max_features= 'sqrt', splitter= 'best')
clf2.fit(X_train, y_train)
```

```python
y_pred=clf2.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Ans 5
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc
```

```python
# Calculate the performance metrics
accuracy=accuracy_score(y_test, y_pred)
precesion=precision_score(y_test, y_pred)
recall=recall_score(y_test, y_pred)
f1=f1_score(y_test, y_pred)
```

```python
# Print the performance metrics
print('Accuracy:', accuracy)
print('Precision:', precesion)
print('Recall:', recall)
print('F1 score:', f1)
```

```python
# Plot of confusion matrix
plot_conf=confusion_matrix(y_pred, y_test)
plot_conf
```

```python
# Calculate the false positive rate and true positive rate for the ROC curve
fpr, tpr, thresholds = roc_curve(y_test, clf2.predict_proba(X_test)[:, 1])

# Calculate the area under the ROC curve
roc_auc = auc(fpr, tpr)
```

```python
# Plot the ROC curve
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Ans 6

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(clf2, filled=True)
plt.show()
```
