```python
# Ans 14
# the top 10 largest apps are

# Remove any duplicate entries based on the 'App' column
df1.drop_duplicates(subset='App', inplace=True)

# Convert installs to numeric values
df1['Installs'] = df1['Installs'].str.replace('[+,]', '', regex=True).astype(int)

# Sort the data by installs in descending order and select the top 10 apps
top_apps = df1.nlargest(10, 'Installs')

# Print the top 10 apps with the highest number of installs
print(top_apps[['App', 'Installs']])
```
